```python
raw_data = pd.read_csv('data/creditcard.csv')
raw_data.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

def standardize_features(data: pd.DataFrame):
    """Standardization for features"""
    scaler = StandardScaler()
    for feature in data.columns[1:-2]:
        data[feature] = scaler.fit_transform(data[feature].values.reshape(-1, 1))

    return data
```

```python
data = standardize_features(raw_data).drop(['Time'], axis=1)
data.head()
```

```python
data.to_csv('data/data.csv', index=None)
```
